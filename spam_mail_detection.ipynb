```python
 #impoting pandas library as pd
import pandas as pd
#importing numpys as np
import numpy as np
 #importing nltk library
import nltk
#omporting wordnet from nltk
nltk.download("wordnet")
#importing stemming from nltk and set stemmer language
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer("english")
#importing lemmatizer from nltk
from nltk.stem import WordNetLemmatizer
#importing tdfivectorizer,train_test_split from sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
#importing logistic regression model and accuracy score from sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
#imporing and reading spam mail csv file.
dt = pd.read_csv("spam_ham_dataset.csv")
```

```python
#to display starting 5 values in the csv file.
dt.head()
```

```python
#displaying labels col only.
dt['label']
```

```python
#changing string values to integer values.
dt['label'].map({'ham':0,'spam':1}).astype(int)
```

```python
#tokenization
def tokenizer(text):
  return text.split()
```

```python
#applying tokenizer
dt['text'].apply(tokenizer)
```

```python
#stemming
def stem_it(text):
  return [stemmer.stem(word) for word in text]

dt['text'].apply(stem_it)
```

```python
#lemmatization
lemmatizer = WordNetLemmatizer()
def lemmit_it(text):
  return [lemmatizer.lemmatize(word) for word in text]

dt['text'].apply(lemmit_it)
```

```python
#vectorization
vectorizer = TfidfVectorizer()
x = vectorizer.fit_transform(dt['text'])
```

```python
#splitting data into testing and training data
y = dt['label']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.33, random_state = 42)
```

```python
#predicting outcome using logistic regression model
logistic = LogisticRegression()
logistic.fit(x_train, y_train)
y_pred = logistic.predict(x_test)
```

```python
#confusion matrix of the model
confusion = confusion_matrix(y_test,y_pred)
print(f"confusion matrix \n {confusion}")

#accuracy score
accuracy = accuracy_score(y_test,y_pred)
print("\n\naccuracy_score\nthe accuracy of the model is:",accuracy)
```
